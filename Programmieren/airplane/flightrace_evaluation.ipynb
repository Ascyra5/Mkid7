<h2 style ="color:dodgerblue"> Aufgenommene Daten <h2>

```python
import numpy as np
```

<h3 style ="color:crimson"> Die Flieglojiglos </h3>


```python
fj = [[9,8,8,5,6,5,10,3], [8,3,7,4,7,7,10,6], [7,6,5,4,3,8,10,5], [9,8,5,6,7,6,10,6]]
fj_widths = [5.5,4.05,10,6.7,10.9]
fj_result = sum(fj_widths) / len(fj_widths)
for i in range(len(fj)):
    fj[i] = np.array(fj[i])
    fj_result += sum(fj[i])
print(fj_result)
print(sum(fj_widths)/(len(fj_widths)))

```

<h3 style ="color:dodgerblue"> The 67-Jet </h3>

```python
j67        = [[5,6,6,7,7,5,10,5], [5,6,2,8,8,9,10,6], [3,4,1,7,8,5,10,8], [4,4,1,8,6,8,10,10], [3,6,2,5,9,8,10,6]]
j67_widths = [4.8,9.9,9.3,10.4,9.9]
j67_result = sum(j67_widths) / len(j67_widths)
for i in range(len(j67)):
    j67[i] = np.array(j67[i])
    j67_result += sum(j67[i])
print(j67_result)
print(sum(j67_widths)/(len(j67_widths)))
```

<h3 style ="color:yellowgreen"> R-Flügler </h3>

```python
rf        = [[6,8,7,8,7,5,10,7,], [5,8,7,9,8,5,10,5], [6,7,5,6,7,7,10,6], [7,6,5,6,7,5,10,5], [7,6,6,6,7,7,10,7]]
rf_widths = [6.8,7.03,8.2,6.3,9.9]
rf_result = sum(rf_widths) / len(rf_widths)
for i in range(len(rf)):
    rf[i] = np.array(rf[i])
    rf_result += sum(rf[i])
print(rf_result)
print(sum(rf_widths)/(len(rf_widths)))
```

<h3 style ="color:orange"> Fliegerle </h3>

```python
fl        = [[7,8,7,9,5,5,10,2], [7,8,8,9,8,9,10,4], [7,7,7,9,8,5,10,6], [9,9,9,10,9,5,10,5], [8,9,8,9,9,9,10,6]]
fl_widths = [9.4,9.3,7.3,8.7,10.9]
fl_result = sum(fl_widths) / len(fl_widths)
for i in range(len(fl)):
    fl[i] = np.array(fl[i])
    fl_result += sum(fl[i])
print(fl_result)
print(sum(fl_widths)/(len(fl_widths)))
```

<h2 style ="color:dodgerblue"> Weitester Flug <h2>

```python
global_max=max(max(fj_widths),max(j67_widths),max(rf_widths),max(fl_widths))
if global_max in fj_widths:
    name = "Flieglojiglos"
elif global_max in j67_widths:
    name = "The 67-Jet"
elif global_max in rf_widths:
    name = "R-Flügler"
else:
    name = "Fliegerle"

print("Die maximale Flugweite ist", global_max, "Meter und wurde von", name, "erreicht.")
```
